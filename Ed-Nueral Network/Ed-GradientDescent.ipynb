```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('insurance_data.csv');
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train , y_test = train_test_split(df[['age','affordibility']], 
                                                      df.bought_insurance, test_size = 0.2,random_state = 25)
```

```python
x_train_s = x_train.copy()
x_train_s['age'] = x_train_s['age'] / 100

x_test_s = x_test.copy()
x_test_s['age'] = x_test_s['age'] / 100
x_train_s.head()
```

```python
model = keras.Sequential([
    keras.layers.Dense(1,input_shape = (2,),activation = 'sigmoid',kernel_initializer = 'ones',bias_initializer = 'zeros')])

model.compile(optimizer = 'adam',
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
```

```python
model.fit(x_train_s, y_train, epochs = 5000)
```

```python
model.evaluate(x_test_s, y_test)
```

```python
model.predict(x_test_s)
```

```python
import math
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
sigmoid(18)
```

```python
def prediction_function(age,affordibility):
    weighted_sum = coef[0] * age + coef[1] * affordibility + intercept
    return sigmoid(weighted_sum)
```

```python
epsilon = 1e-15
def log_loss(y_predicted , y_true):
    y_predicted_new = np.clip(y_predicted, epsilon, 1-epsilon).astype(np.float64)
    return -np.mean(y_true * np.log(y_predicted_new) + (1 - y_true) * np.log(1 - y_predicted_new))
```

```python
def sigmoid_numpy(X):
    return 1 / (1 + np.exp(-X))
sigmoid_numpy(np.array([12,0,1]))
```

```python
def gradient_descent(age,affordibility, y_true, epochs, loss_threshold):
    # w1, w2, bias
    w1 = w2 = 1
    bias = 0
    rate = 0.5
    n = len(age)
    for i in range(epochs):
        weighted_sum = w1 * age + w2 * affordibility + bias
        y_predicted = sigmoid_numpy(weighted_sum)

        loss = log_loss(y_predicted, y_true)

        w1d = (1/n) * np.dot(np.transpose(age) , (y_predicted - y_true))
        w2d = (1/n) * np.dot(np.transpose(affordibility) , (y_predicted - y_true))
        
        bias_d = np.mean(y_predicted - y_true)

        w1 = w1 - rate *w1d
        w2 = w2 - rate *w2d

        bias = bias - rate * bias_d

        print(f'Epoch:{i}, w_age:{w1}, w_affordability:{w2}, bias:{bias}, loss:{loss}')
        if loss <= loss_threshold:
            break
    return w1,w2, bias
```
