```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("data/customer-churn.csv")
df.sample(5)
```

```python
df.drop('customerID',axis = 'columns',inplace=True)
df.dtypes
```

```python
df.TotalCharges.values
```

```python
pd.to_numeric(df.TotalCharges, errors = 'coerce').isnull()
```

```python
df[pd.to_numeric(df.TotalCharges, errors = 'coerce').isnull()]
```

```python
print("Total nulls: " ,df[pd.to_numeric(df.TotalCharges, errors = 'coerce').isnull()].shape)
print("Total: " ,df.shape)
```

```python
df = df[df.TotalCharges != ' ']
df.shape
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges)
df.TotalCharges
```

```python
df.TotalCharges .dtypes
```

```python
tenure_Churn_No = df[df.Churn == 'No'].tenure
tenure_Churn_Yes = df[df.Churn == 'Yes'].tenure

plt.hist([tenure_Churn_Yes, tenure_Churn_No], color=['green', 'red'], label=['Churn Yes', 'Churn No'])
plt.legend()
plt.xlabel('Tenure')
plt.ylabel('Frequency')
plt.title('Distribution of Tenure for Churned and Non-Churned Customers')
plt.show()
```

```python
monthly_Churn_No = df[df.Churn == 'No'].MonthlyCharges
monthly_Churn_Yes = df[df.Churn == 'Yes'].MonthlyCharges

plt.hist([monthly_Churn_Yes, monthly_Churn_No], color=['green', 'red'], label=['Churn Yes', 'Churn No'])
plt.legend()
plt.xlabel('MonthlyCharges')
plt.ylabel('Frequency')
plt.title('Distribution of Tenure for Churned and Non-Churned Customers')
plt.show()
```

```python
def shwing_unique_col_values(df, numeric):
    for col in df:
        if numeric == False:
            if df[col].dtype == 'object':
                print(f' {col} : {df[col].unique()}')
        else:
            print(f' {col} : {df[col].unique()}')
```

```python
shwing_unique_col_values(df,False)
```

```python
df.replace('No internet service','No',inplace = True)
df.replace('No phone service','No',inplace = True)
shwing_unique_col_values(df,False)
```

```python
binary_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection'
                  ,'TechSupport' ,'StreamingTV','StreamingMovies','PaperlessBilling','Churn']

for col in binary_columns:
    df[col].replace({'Yes' : 1, 'No' : 0},inplace = True)
```

```python
shwing_unique_col_values(df,True)
```

```python
df['gender'].replace({'Male' : 1, 'Female' : 0},inplace = True)
df.dtypes
```

```python
df = pd.get_dummies(data= df,columns=['InternetService','Contract','PaymentMethod'])
```

```python
df.dtypes
```

```python
cat_binary_columns = ['InternetService_DSL','InternetService_Fiber optic','InternetService_No',
                  'Contract_Month-to-month','Contract_One year','Contract_Two year','PaymentMethod_Bank transfer (automatic)'
                  ,'PaymentMethod_Credit card (automatic)' ,'PaymentMethod_Electronic check','PaymentMethod_Mailed check']

for col in cat_binary_columns:
    df[col].replace({True : 1, False : 0},inplace = True)
```

```python
df.dtypes
```

```python
df.head()
```

```python
scale_cols = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler()

df[scale_cols] = scaler.fit_transform(df[scale_cols])
```

```python
df.head()
```

```python
x = df.drop('Churn',axis = 'columns')
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 5)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
        keras.layers.Dense(20, input_shape = (26,),activation = 'relu'),
        keras.layers.Dense(1, activation = 'sigmoid'),

])


model.compile(optimizer = 'adam',
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train,epochs = 150)
```

```python
model.evaluate(x_test, y_test)
```
